# Building a Simple Neural Network

## 1) Import Libraries

```python
import torch
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm
```

## 2) Defining data

```python
# distance in miles
distances = torch.tensor([[1.0], [2.0], [3.0], [4.0]], dtype=torch.float32)

# delivery times in minutes
times = torch.tensor([[6.96], [12.11], [16.77], [22.21]], dtype=torch.float32)
```

## 3) Model Building

```python
model = nn.Sequential(nn.Linear(1, 1))
```

```python
# Define the loss function and optimizer
loss_function = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

## 4) Training

```python
for epoch in tqdm(range(500)):
    # 1st step: reset the optimizer
    optimizer.zero_grad()

    # 2nd step: make predictions
    outputs = model(distances)

    # 3rd step: calculate the loss - how bad was this guess?
    loss = loss_function(outputs, times)

    # 4th step: calculate adjustments
    loss.backward()

    # 5th step: update the model
    optimizer.step()

    # print loss every 50 epochs
    if (epoch + 1) % 50 == 0:
        print(f"Epoch {epoch + 1}: Loss = {loss.item()}")
```

## 5) Inference

```python
with torch.no_grad():
    test_distance = torch.tensor([[25.0]], dtype=torch.float32)
    predicted_time = model(test_distance)
    print(f"Predicted delivery time for {test_distance.item()} miles: {predicted_time.item():.2f} minutes")
```
