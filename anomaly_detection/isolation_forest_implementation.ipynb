# Isolation Forest Anomaly Detection

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/healthcare.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df.iloc[:, 0], df.iloc[:, 1])
```

```python
from sklearn.ensemble import IsolationForest
```

```python
clf_model = IsolationForest(contamination=0.2)
clf_model.fit(df)
predictions = clf_model.predict(df)
```

```python
predictions
```

```python
import numpy as np

index_outliers = np.where(predictions < 0)
index_outliers
```

```python
x = df.values
```

```python
plt.scatter(df.iloc[:, 0], df.iloc[:, 1])
plt.scatter(x[index_outliers, 0], x[index_outliers, 1], edgecolors="r")
```
