# Introdução a testes

## 1. Carregamento de bibliotecas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2. Carregamento das bases de dados - TMDB

```python
tmdb = pd.read_csv('../data/tmdb_5000_movies.csv')

tmdb.head()
```

## 3. Análise exploratória da base de dados - TMDB

```python
tmdb.describe()
```

```python
sns.histplot(tmdb['vote_average'], kde=True, bins=20, stat="density")
plt.grid(True, alpha=0.3)
plt.xlabel("Nota Média")
plt.ylabel("Densidade")
plt.title("Média de votos em filmes no TMDB 5000")
plt.show()
```

```python
sns.boxplot(
    tmdb["vote_average"]
)

plt.grid(True, alpha=.3)
plt.title("Distribuição de notas médias dos filmes do TMDB 5000")

plt.ylabel("Distribuição das notas médias")

plt.show()
```

### 3.1. Filtro dos filmes com 10 ou mais votos

```python
tmdb_filtered = tmdb.query("vote_count >= 10")
```

```python
tmdb_filtered.describe()
```

```python
sns.histplot(tmdb_filtered['vote_average'], kde=True, bins=20)
plt.grid(True, alpha=0.3)
plt.xlabel("Nota Média")
plt.ylabel("Frequência")
plt.title("Média de votos em filmes no TMDB 5000 com 10 ou mais votos")
plt.show()
```

```python
sns.boxplot(
    tmdb_filtered["vote_average"]
)

plt.grid(True, alpha=.3)
plt.title("Distribuição de votos em filmes no TMDB 5000 com 10 ou mais votos")

plt.ylabel("Distribuição das notas médias")

plt.show()
```

## 4. Carregamento das bases de dados - MovieLens

```python
notas_movieLens = pd.read_csv("../data/ratings.csv")
nota_media_por_filme = notas_movieLens.groupby("movieId").agg(
    {
        "rating": ["mean", "count"],

    }
)
nota_media_por_filme.columns = ["media", "quantidade_votos"]

nota_media_por_filme = nota_media_por_filme.query("quantidade_votos >= 10")

nota_media_por_filme.head()
```

```python
sns.histplot(nota_media_por_filme["media"], kde=True, bins=20)
plt.grid(True, alpha=0.3)
plt.xlabel("Nota Média")
plt.ylabel("Frequência")
plt.title("Média de votos em filmes no MovieLens 100k com 10 ou mais votos")
plt.show()
```

```python
sns.boxplot(
    nota_media_por_filme["media"]
)

plt.grid(True, alpha=.3)
plt.title("Distribuição de votos em filmes no MovieLens com 10 ou mais votos")

plt.ylabel("Distribuição das notas médias")

plt.show()
```

## 5. Distribuição acumulativa

### 5.1. TMDB 5000

```python
sns.histplot(tmdb_filtered["vote_average"], kde=True, bins=20, cumulative=True, stat="percent")
plt.grid(True, alpha=0.3)
plt.xlabel("Nota Média")
plt.ylabel("Frequência")
plt.title("Média de votos em filmes no TMDB 5000 com 10 ou mais votos")
plt.show()
```

### 5.2. MediaLens 100k

```python
sns.histplot(nota_media_por_filme["media"], kde=True, bins=20, cumulative=True, stat="percent")
plt.grid(True, alpha=0.3)
plt.xlabel("Nota Média")
plt.ylabel("Frequência")
plt.title("Média de votos em filmes no MovieLens 100k com 10 ou mais votos")
plt.show()
```

## 6. Análises a distribuição de outros campos do TMDB

### 6.1. Número de votos

```python
sns.histplot(tmdb_filtered["vote_count"], kde=True, bins=20)
plt.grid(True, alpha=0.3)
plt.xlabel("Quantidade de Votos")
plt.ylabel("Frequência")
plt.title("Quantidade de votos em filmes no TMDB 5000 com 10 ou mais votos")
plt.show()
```

### 6.2. Orçamento do filme

```python
sns.histplot(tmdb.query("budget > 0")["budget"], kde=True, bins=20)
plt.grid(True, alpha=0.3)
plt.xlabel("Orçamento do filme")
plt.ylabel("Frequência")
plt.title("Orçamento dos filmes no TMDB 5000")
plt.show()
```

### 6.3. Popularidade do filme

```python
sns.histplot(tmdb["popularity"], kde=True, bins=20)
plt.grid(True, alpha=0.3)
plt.xlabel("Popularidade do filme")
plt.ylabel("Frequência")
plt.title("Popularidade dos filmes no TMDB 5000")
plt.show()
```

### 6.4. Tempor de duração do filme

```python
sns.histplot(tmdb.query("runtime > 0").dropna()["runtime"], kde=True, bins=20)
plt.grid(True, alpha=0.3)
plt.xlabel("Tempo de duração do filme")
plt.ylabel("Frequência")
plt.title("Tempo de duração dos filmes no TMDB 5000")
plt.show()
```

```python
sns.histplot(tmdb.query("runtime > 0")["runtime"].dropna(), kde=True, bins=20, cumulative=True, stat="percent")
plt.grid(True, alpha=0.3)
plt.xlabel("Tempo de duração do filme")
plt.ylabel("percentual de filmes")
plt.title("Tempo de duração dos filmes no TMDB 5000")
plt.show()
```
